```python
import cv2
import numpy
import matplotlib.pyplot as plt
from scipy import signal
img = cv2.imread(r"C:\Users\nandita\Desktop\lena.png",0)
```

```python
plt.imshow(img)
n1,n2 = img.shape
plt.imsave("original.jpg", img)
```

```python
# extracting the input image
input_img = cv2.resize(img,(n1//2,n2//2))
r_input ,c_input = input_img.shape
#upscale/zoom input_image
scaled_image = np.zeros((2*r_input,2*c_input))
scaled_image[0::2,0::2]= input_img
```

```python
#interpolation using Rectangular window
import skimage
for i in range(2,9):
    w = signal.windows.boxcar(i)
    w2d = np.sqrt(np.outer(w,w))
    filtered = signal.convolve(scaled_image,w2d, mode='same') 
    plt.imsave(str(i)+"rec.jpg",filtered)
    plt.imsave(str(i)+"diff_rec.jpg", abs(img-filtered))
    psnr = skimage.metrics.peak_signal_noise_ratio(img,filtered)
    ssim = skimage.measure.compare_ssim(img,filtered)
    print("PSNR= ",psnr, end=" ")
    print("SSIM= ", ssim)
    


```

```python
#interpolation using Hanning window
for i in range(2,9):
    w = signal.windows.hann(i)
    w2d = np.sqrt(np.outer(w,w))
    filtered = signal.convolve(scaled_image,w2d, mode='same') 
#     plt.imsave(str(i)+"hann.jpg",filtered)
#     plt.imsave(str(i)+"diff_hann.jpg", abs(img-filtered))
#     plt.figure()
#     plt.imshow(filtered)
    psnr = skimage.metrics.peak_signal_noise_ratio(img,filtered)
    ssim = skimage.measure.compare_ssim(img,filtered)
    print("PSNR= ",psnr, end=" ")
    print("SSIM= ", ssim)
```

```python
#interpolation using Hamming window
for i in range(2,9):
    w = signal.windows.hamming(i)
    w2d = np.sqrt(np.outer(w,w))
    filtered = signal.convolve(scaled_image,w2d, mode='same') 
#     plt.imsave(str(i)+"hamming.jpg",filtered)
#     plt.imsave(str(i)+"diff_hamming.jpg", abs(img-filtered))
#     plt.figure()
#     plt.imshow(filtered)
    psnr = skimage.metrics.peak_signal_noise_ratio(img,filtered)
    ssim = skimage.measure.compare_ssim(img,filtered)
    print("PSNR= ",psnr, end=" ")
    print("SSIM= ", ssim)
    
```

```python
#interpolation using blackmanharris window
for i in range(2,9):
    w = signal.windows.blackmanharris(i)
    w2d = np.sqrt(np.outer(w,w))
    filtered = signal.convolve(scaled_image,w2d, mode='same') 
    plt.imsave(str(i)+"bh.jpg",filtered)
    plt.imsave(str(i)+"diff_bh.jpg", abs(img-filtered))
    plt.figure()
    plt.imshow(filtered)
    psnr = skimage.metrics.peak_signal_noise_ratio(img,filtered)
    ssim = skimage.measure.compare_ssim(img,filtered)
    print("PSNR= ",psnr, end=" ")
    print("SSIM= ", ssim)
    
```

```python
# upscaling and comparing with different inbuilt interpolation types
types = ["nearest", "linear", "cubic", "area", "lanczos"]
nearest, linear, cubic, area, lanczos=  (
        cv2.resize(input_img, (256,256), interpolation=interp)
        for interp in [
            cv2.INTER_NEAREST,
            cv2.INTER_LINEAR,
            cv2.INTER_CUBIC,
            cv2.INTER_AREA,
            cv2.INTER_LANCZOS4
        ]
    )
#Metrics for different interpolations
l= [nearest, linear, cubic, area, lanczos]
for i in range(5):
    psnr = skimage.metrics.peak_signal_noise_ratio(l[i],img)
    ssim = skimage.measure.compare_ssim(img,l[i])
    print("For "+ types[i]+": "+"PSNR= ",psnr, end=" ")
    print("SSIM= ", ssim)

figsize = [6, 16] 
fig, ax = plt.subplots(nrows=5, ncols=2,figsize= figsize)

# plot simple raster image on each sub-plot
for i, axi in enumerate(ax.flat):
    output = l[i//2]
    if i%2==0:
        axi.imshow(output)
        rowid = i // ncols
        colid = i % ncols
        axi.set_title(types[i//2])
    else:
        difference = img-output
        axi.imshow(difference)
        rowid = i // ncols
        colid = i % ncols
        axi.set_title("Difference")

plt.show()
#fig.savefig("inbuilt.jpg")
```

```python
# Algorithm for Bilinear interpolation

def interpolate_bilinear(array_in, width_in = 128, height_in =128):
    height_out = 256
    width_out = 256
    array_out = np.zeros((256,256))
    for i in range(height_out):
        for j in range(width_out):
            # Relative coordinates of the pixel in output space
            x_out = j / width_out
            y_out = i / height_out
 
            # Corresponding absolute coordinates of the pixel in input space
            x_in = (x_out * width_in)
            y_in = (y_out * height_in)
 
            # Nearest neighbours coordinates in input space
            x_prev = int(np.floor(x_in))
            x_next = x_prev + 1
            y_prev = int(np.floor(y_in))
            y_next = y_prev + 1
 
            # Sanitize bounds - no need to check for < 0
            x_prev = min(x_prev, width_in - 1)
            x_next = min(x_next, width_in - 1)
            y_prev = min(y_prev, height_in - 1)
            y_next = min(y_next, height_in - 1)
            
            # Distances between neighbour nodes in input space
            Dy_next = y_next - y_in;
            Dy_prev = 1. - Dy_next; # because next - prev = 1
            Dx_next = x_next - x_in;
            Dx_prev = 1. - Dx_next; # because next - prev = 1
            
            # Interpolate 
            array_out[i][j] = Dy_prev * (array_in[y_next][x_prev] * Dx_next + array_in[y_next][x_next] * Dx_prev) \
                + Dy_next * (array_in[y_prev][x_prev] * Dx_next + array_in[y_prev][x_next]* Dx_prev)
                
    return array_out

filtered = interpolate_bilinear(input_img, c_input, r_input)
# plt.imsave("bilinear.png", filtered)
# plt.imsave("diff_billinear.png", abs(img-filtered))
psnr = skimage.metrics.peak_signal_noise_ratio(img,filtered)
ssim = skimage.measure.compare_ssim(img,filtered)
print("PSNR= ",psnr, end=" ")
print("SSIM= ", ssim)
```
